```python
litigations['Company'] = litigations['Company'].str.lower()
firm_code['conm'] = firm_code['conm'].str.lower()
litigations['Parent Company'] = litigations['Parent Company'].str.lower()
```

```python
def is_similar(item, code):
  counter = 0
  item_list = word_tokenize(item)
  try:
    item_list.remove('.')
  except:
    pass
  try:
    item_list.remove(',')
  except:
    pass
  code_list = word_tokenize(code)
  try:
    code_list.remove('.')
  except:
    pass
  try:
    code_list.remove(',')
  except:
    pass
  for word in item_list:
    if word in code_list:
      counter += 1
  if counter/len(code_list) >= 0.6:
    return True
  else:
    return False
```

```python
conm_code = firm_code.conm.unique()
comp_code = litigations.Company.unique()
comp2_code = litigations['Parent Company'].unique()[1:]
```

```python
litigations['conm'] = np.nan
litigations['conm2'] = np.nan
for item in comp_code[200:300]:
  print(f'item = {item}')
  conm_list = []
  for jtem in conm_code:
    if is_similar(item, jtem):
      conm_list.append(jtem)
  litigations.conm[litigations.Company == item] = str(conm_list)
for ktem in comp2_code[200:300]:
  conm2_list = []
  for jtem in conm_code:
    if is_similar(ktem, jtem):
      conm2_list.append(jtem)
  litigations.conm2[litigations['Parent Company'] == ktem] = str(conm2_list)
litigations.to_csv('litigations_w_conm1_conm2_200_300.csv')
```
