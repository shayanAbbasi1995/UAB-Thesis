```python
import pandas as pd
import numpy as np
```

```python
lit100 = pd.read_csv('litigations_w_conm1_conm2_0_100.csv')
lit200 = pd.read_csv('litigations_w_conm1_conm2_100_200.csv')
lit300 = pd.read_csv('litigations_w_conm1_conm2_200_300.csv')
lit400 = pd.read_csv('litigations_w_conm1_conm2_300_400.csv')
lit500 = pd.read_csv('litigations_w_conm1_conm2_400_500.csv')
lit1000 = pd.read_csv('litigations_w_conm1_conm2_500_1000.csv')
lit1500 = pd.read_csv('litigations_w_conm1_conm2_1000_1500.csv')
lit2000 = pd.read_csv('litigations_w_conm1_conm2_2000.csv')
```

```python
def none_emp_lst(df):
  for i in range(len(df.conm)):
    if df.conm.iloc[i] == '[]':
      df.conm.iloc[i] = None
    if df.conm2.iloc[i] == '[]':
      df.conm2.iloc[i] = None
  return df
```

```python
lit100 = none_emp_lst(lit100)
lit200 = none_emp_lst(lit200)
lit300 = none_emp_lst(lit300)
lit400 = none_emp_lst(lit400)
lit500 = none_emp_lst(lit500)
lit1000 = none_emp_lst(lit1000)
lit1500 = none_emp_lst(lit1500)
lit2000 = none_emp_lst(lit2000)
```

```python
lit100.conm.fillna(lit200.conm, inplace=True)
lit100.conm.fillna(lit300.conm, inplace=True)
lit100.conm.fillna(lit400.conm, inplace=True)
lit100.conm.fillna(lit500.conm, inplace=True)
lit100.conm.fillna(lit1000.conm, inplace=True)
lit100.conm.fillna(lit1500.conm, inplace=True)
lit100.conm.fillna(lit2000.conm, inplace=True)


lit100.conm2.fillna(lit200.conm2, inplace=True)
lit100.conm2.fillna(lit300.conm2, inplace=True)
lit100.conm2.fillna(lit400.conm2, inplace=True)
lit100.conm2.fillna(lit500.conm2, inplace=True)
lit100.conm2.fillna(lit1000.conm2, inplace=True)
lit100.conm2.fillna(lit1500.conm2, inplace=True)
lit100.conm2.fillna(lit2000.conm2, inplace=True)
```

```python
final_lit = lit100[lit100['Ownership Structure'] == 'publicly traded']
final_lit = none_emp_lst(final_lit)
final_lit
```
